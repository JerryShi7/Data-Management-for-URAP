```python
import csv, pandas
```

```python
#initialize some helper lists and var names
fileList = ['4col98f.csv', '4col99f.csv', '4col00f.csv', '4col01f.csv', '4col02f.csv', '4col03f.csv', '4col04f.csv',\
             '4col05f.csv', '4col06f.csv','4col07f.csv', '4col08f.csv', '4col09f.csv']
fileList2 = ['6col10f.csv', '6col11f.csv', '6col12f.csv', '6col13f.csv']
bufList = []
file14 = '14.csv'
file15 = '15.csv'
file16 = '16.csv'
```

```python
#read 15 and 16 files
buf15 = pandas.read_csv(file15, encoding = 'latin1')
buf16 = pandas.read_csv(file16, encoding = 'latin1')
```

```python
#extract columns we need and rename them to make the column names cosistent
buf15new = buf15[['NCESSCH', 'SCH_NAME', 'CHARTER_TEXT', 'SY_STATUS', 'RECON_STATUS']]
buf16new = buf16[['NCESSCH', 'SCH_NAME', 'CHARTER_TEXT', 'SY_STATUS', 'recon_status']]
buf15new.rename(columns={'SCH_NAME': 'SCHNAM15', 'CHARTER_TEXT': 'CHARTER_TEXT15', \
                         'SY_STATUS': 'SY_STATUS15', 'RECON_STATUS': 'RECON_STATUS15'}, inplace=True)
buf16new.rename(columns={'SCH_NAME': 'SCHNAM16', 'CHARTER_TEXT': 'CHARTER_TEXT16', \
                         'SY_STATUS': 'SY_STATUS16', 'recon_status': 'RECON_STATUS16'}, inplace=True)
```

```python
#read all the files from 98 to 09 and put them into bufList
for i in range(0, 12):
   bufList.append(pandas.read_csv(fileList[i], encoding = 'latin1'))
```

```python
#delete the strange indexes from the 98 to 09 bufs and put them into a new bufList
newbufList = []
for buf in bufList:
    temp = buf[buf.columns[2:6]]
    newbufList.append(temp)

#read 10 - 13 files which do not have indexes, and rename columns in order to be consistent
buf10 = pandas.read_csv(fileList2[0], encoding = 'latin1')
buf11 = pandas.read_csv(fileList2[1], encoding = 'latin1')
buf12 = pandas.read_csv(fileList2[2], encoding = 'latin1')
buf13 = pandas.read_csv(fileList2[3], encoding = 'latin1')
buf10.rename(columns={'SCHNAM': 'SCHNAM10', 'STATUS': 'STATUS10', 'CHARTR': 'CHARTR10','RECONSTF': 'RECONSTF10',\
                     'RECONSTY': 'RECONSTY10'}, inplace=True)
buf11.rename(columns={'SCHNAM': 'SCHNAM11', 'STATUS': 'STATUS11', 'CHARTR': 'CHARTR10','RECONSTF': 'RECONSTF10',\
                     'RECONSTY': 'RECONSTY10'}, inplace=True)
buf12.rename(columns={'SCHNAM': 'SCHNAM12', 'STATUS': 'STATUS12', 'CHARTR': 'CHARTR10','RECONSTF': 'RECONSTF10',\
                     'RECONSTY': 'RECONSTY10'}, inplace=True)
buf13.rename(columns={'SCHNAM': 'SCHNAM13', 'STATUS': 'STATUS13', 'CHARTR': 'CHARTR10','RECONSTF': 'RECONSTF10',\
                     'RECONSTY': 'RECONSTY10'}, inplace=True)

#add 10 - 13, 15 - 16 buffers into the new bufList
newbufList.append(buf10)
newbufList.append(buf11)
newbufList.append(buf12)
newbufList.append(buf13)
newbufList.append(buf15new)
newbufList.append(buf16new)
```

```python
#read the big 14 file
df14 = pandas.read_csv(file14, encoding = 'latin1')
```

```python
# start to merge 
for buf in newbufList:
    temp = pandas.merge(df14, buf, how = 'outer', on = ['NCESSCH'])
    print("merging...")
    df14 = temp
print(df14)
```

```python
#check duplicates
print(len(df14['NCESSCH']) == len(df14['NCESSCH'].unique()), len(df14['NCESSCH']), len(df14['NCESSCH'].unique()), " unique for file 1")
```

```python
#convert to csv
df14.to_csv('final.csv', index = False)
```
