```python
import pandas as pd
```

```python
df1 = pd.read_csv("withPVI_final.csv", encoding = "latin1")
```

```python
df2 = pd.read_csv("geo.csv", encoding = "latin1")
df3 = df2[['NCESSCH','LATCODE', 'LONGCODE','CONUM','CONAME',\
          'CD','LOCALE', 'CBSA', 'CSA', 'NECTA','METMIC']]
```
